```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
```

```python
# Preprocess data
df = pd.read_csv('store_data.csv', header=None)
df = df.astype(str)
df = df.dropna()
```

```python
# Change apriori output to a dataframe
def visualize_rules_2(list_of_results):
    values = []
    for result in list_of_results:
        statistics = []
        # Find item - Append item to statistics list
        primary_behavior = tuple(result[2][0][0])[0]
        statistics.append(primary_behavior)

        consequent_behavior = tuple(result[2][0][1])[0]
        statistics.append(consequent_behavior)

        support = result[1]
        statistics.append(support)

        confidence = result[2][0][2]
        statistics.append(confidence)

        lift = result[2][0][3]
        statistics.append(lift)

        values.append(statistics)

    dataframe = pd.DataFrame(values, columns = ["Primary Product", "Consequent Product", "Support", "Confidence", "Lift"])
    return dataframe
```

```python
from apyori import apriori

rules = apriori( 
  transactions = df.values,
  min_support = 0.01,
  min_confidence = 0.2,
  min_lift = 2,
  )

results = list(rules)
```

```python
df_results = visualize_rules_2(results)
df_results = df_results.drop_duplicates()
df_results.nlargest(n = 10, columns = ["Lift"])
```
