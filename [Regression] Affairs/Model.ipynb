```python
# Data Preprocessing

import pandas as pd

dataframe = pd.read_csv("Affairs.csv")
dataframe = dataframe.drop(dataframe.columns[-1], axis = 1)

# Change children to a binary numeric column
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
children_column = label_encoder.fit_transform(dataframe["children"].values)
dataframe["children"] = children_column.reshape(len(dataframe), 1)

# Change gender into 2 columns "male - female"
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

column_transformer = ColumnTransformer(transformers = [('encoder', OneHotEncoder(), [1])], remainder = 'passthrough')
new_values = column_transformer.fit_transform(dataframe.values)

# Create a new dataframe
dataframe = pd.DataFrame(new_values, columns = ['female', 'male', 'affairs', 'age', 'years_married', 'has_children', 'is_religious', 'education', 'occupation', 'relationship_rating'])
dataframe
```

```python
import arff

with open('Affairs.arff', 'w') as f:
    arff.dump({
        'data': dataframe.values.tolist(),
        'attributes': [(col, 'NUMERIC') for col in dataframe.columns],
        'relation': 'Affairs'
    }, f)
```
